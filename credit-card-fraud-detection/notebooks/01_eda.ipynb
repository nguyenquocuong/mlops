```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/creditcard.csv")
df.head()
```

# 1. Class imbalance

```python
print(df["Class"].value_counts())
print(f"Fraud rate: {df['Class'].mean()*100:.3f}")
```

# 2. Amount distribution by class

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
df[df["Class"] == 0]["Amount"].hist(bins=50, ax=axes[0], label="Legit", alpha=0.7)
df[df["Class"] == 1]["Amount"].hist(bins=50, ax=axes[1], label="Fraud", alpha=0.7, color="red")
axes[0].set_title("Legit Transaction Amounts")
axes[1].set_title("Fraud Transaction Amounts")
plt.tight_layout()
```

```python
# 3. Time distribution — frauds cluster at certain hours?
df["Hour"] = (df["Time"] % 86400) // 3600
fraud_by_hour = df.groupby(["Hour", "Class"]).size().unstack(fill_value=0)
fraud_by_hour.plot(kind="bar", figsize=(14, 4), title="Transactions by Hour")
```

```python
# 4. Feature correlation with Class
corr = df.corr()["Class"].drop("Class").sort_values()
corr.plot(kind="barh", figsize=(8, 10), title="Feature Correlation with Fraud")
plt.tight_layout()
```
